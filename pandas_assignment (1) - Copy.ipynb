## 1. Create a DataFrame with the following data:  

```python
import pandas as pd
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]}
```

```python
df = pd.DataFrame(data)
print(df.head())
```

```python
print(df[['Age','Salary']].describe())
```

```python
hr_employees = df[df['Department'] == 'HR']
avg_salary_hr = hr_employees['Salary'].mean()

print(f"Average salary in HR department: ${avg_salary_hr:.2f}")
```

```python
df['Bonus'] = df['Salary']*0.1
print(df)
```

```python
filter_age = df[(df['Age']>=25)&(df['Age']<=30)]
print(filter_age)
```

```python
dept_avg_salary = df.groupby('Department')['Salary'].mean()
print(dept_avg_salary)
```

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
print(sorted_df)
sorted_df.to_csv('employees_sorted_by_salary.csv', index=False)
```
