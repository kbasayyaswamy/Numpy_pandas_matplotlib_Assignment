## Practice data visualization techniques for better data representation.

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x,y,marker='o',label="plot chat")
plt.title("simple line plot")
plt.xlabel(" X valves")
plt.ylabel(" Y valves")
plt.legend()
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
c = ['red','green','blue']
plt.bar(students,marks,color = c)
plt.xlabel('Student')
plt.ylabel('Marks')
plt.title('Student performance')
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
plt.figure(figsize=(6, 6))
explode = [0.1 if value == max(revenue) else 0 for value in revenue]
plt.pie(revenue,labels=regions,explode=explode,autopct='%1.1f%%',startangle=140,shadow=True)

plt.title("Percentage Distribution of Revenue by Region")
plt.axis('equal')  
plt.show()
```

```python
import numpy as np
data = np.random.randint(1, 101, 1000)
plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, edgecolor='black')

plt.title("Frequency Distribution of Random Integers (1100)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
