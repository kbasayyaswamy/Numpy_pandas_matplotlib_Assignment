## 1.Create a 1D NumPy array with integers from 1 to 20. Perform the following operations

```python
import numpy as np
arr = np.arange(1,21)
print(arr)
sum = np.sum(arr)
mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)
print("\n",sum,"\n",mean,"\n",median,"\n",std_dev)
```

```python
indices = [ i for i in range (len(arr)) if arr[i] > 10]
print(arr[indices])
```

## 3.Create two 3 X 3 arrays filled with random integers between 1 and 20

```python
arr3 = np.random.randint(1,21,(3,3))
arr4 = np.random.randint(1,21,(3,3))
arr3,arr4
```

```python
addition = arr3 + 5
print("addition")
print(addition,"\n")
mul = arr3 * 5
print("Multiplication")
print(mul,"\n")
sub = arr3 - 5
print("Subtraction")
print(sub,"\n")
dot_product = np.dot(arr3,arr4)
print("dot_product")
print(dot_product)
```

## 4.Reshape a 1D array of size 12 into a 3 X 4 2D array and slice the first two rows and last two columns.

```python
arr4 = np.arange(12)
reshaped_arr = arr4.reshape(3, 4)
print("Reshaped 3x4 array:")
print(reshaped_arr)
sliced_arr = reshaped_arr[0:2, 2:4]
print("\nSliced array (first two rows and last two columns):")
print(sliced_arr)
```
